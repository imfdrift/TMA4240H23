```python
import numpy as np
import matplotlib.pyplot as plt

x = [850,740,900,1070,930,850,950,980,980,880,1000,980,930,650,760,810,
     1000,1000,960,960,960,940,960,940,880,800,850,880,900,840,830,790,
     810,880,880,830,800,790,760,800,880,880,880,860,720,720,620,860,970,
     950,880,910,850,870,840,840,850,840,840,840,890,810,810,820,800,770,
     760,740,750,760,910,920,890,860,880,720,840,850,850,780,890,840,780,
     810,760,810,790,810,820,850,870,870,810,740,810,940,950,800,810,870]

plt.hist(x)
plt.xlabel('speed')
plt.ylabel('number of observations')
plt.title('Michelson data')
plt.show()
```

```python
z = np.random.normal(size=100)
plt.hist(z)
plt.xlabel('value')
plt.ylabel('number of observations')
plt.title('Simulated normal data')
plt.show()
```

```python
import statsmodels.api as sm
import pylab as pylab
import scipy.stats as stats


# create Q-Q plot 
stats.probplot(x, dist="norm", plot=pylab)
pylab.show()


```

```python
z = np.random.normal(size=100)

# create Q-Q plot 
stats.probplot(z, dist="norm", plot=pylab)
pylab.show()
```

```python
muhat = np.mean(x)
sigma2hat = np.var(x,ddof=0)    # ddof=0 gj√∏r at man ganger med 1/n

print('muhat: ',muhat)
print('sigma2hat: ',sigma2hat)
```

```python
print('hat{Var(muhat)}: ', sigma2hat/len(x))
print('hat{SD(muhat)}: ', np.sqrt(sigma2hat/len(x)))
```

### Gjentar de siste utregningene med forventningsrett estimator for sigma^2.

```python
muhat = np.mean(x)
s2 = np.var(x,ddof=1)    # ddof=0 gj√∏r at man ganger med 1/(n-1)

print('muhat: ',muhat)
print('s2: ',s2)

print('hat{Var(muhat)}: ', s2/len(x))
print('hat{SD(muhat)}: ', np.sqrt(s2/len(x)))
```
